```python
df=pd.read_csv("https://raw.githubusercontent.com/shsarv/Machine-Learning-Projects/main/Diabetes%20Prediction%20%5BEND%202%20END%5D/dataset/kaggle_diabetes.csv")
```

```python
X=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_train,y_train,y_train=train_test_split(X,y,test_size=0.25,random_state=45)
```

```python
lgr.fit(X_train,y_train)
```

```python
lgr.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lgr,X_train,y_train,cv=5)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.01,0.1, 1, 10, 100],  # regularization parameter - Changed 'c' to 'C'
    "penalty": ["l2", "l1"],  # penalty
    "solver": ["liblinear"]  # optimization algorithm
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression # Import LogisticRegression
log_reg = LogisticRegression(max_iter = 1000)
```

```python
grid_search=GridSearchCV(estimator=log_reg,param_grid=param_grid,cv=5,scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
lgr=LogisticRegression(max_iter=1000,C=10,penalty='l2',solver='liblinear')
lgr.fit(X_train,y_train)
```

```python
lgr.score(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,lgr.predict(X_train)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,lgr.predict(X_train))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score

print(accuracy_score(y_train,lgr.predict(X_train)))
print(precision_score(y_train,lgr.predict(X_train)))
print(recall_score(y_train,lgr.predict(X_train)))
print(f1_score(y_train,lgr.predict(X_train)))
```
